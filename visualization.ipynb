## <b>Data Visualization</b>

Data visualization is a powerful technique that helps us understand and communicate complex information in a visual format. Python, with its extensive libraries and tools, has become one of the most popular choices for data visualization. In this comprehensive guide, we will explore the world of data visualization with Python, covering various libraries, techniques, and best practices.

![](https://miro.medium.com/v2/resize:fit:546/0*w7IASLYhX1AUQ9QH)

##### <b> Why Does Data Visualization Matters?</b>

<b> &nbsp; Data visualization matters for several compelling reasons:</b>

- **Visual Perception is Powerful :** Humans are highly visual creatures, and we process visual information more efficiently than textual or numerical data. Visualization leverages our natural ability to interpret patterns and trends quickly, making it easier to understand complex information.
- **Simplifies Complex Data :** Raw data can be overwhelming and challenging to grasp in its original form. Visualization simplifies complex datasets by transforming them into visual representations, making it easier to identify insights and patterns.
- **Spotting Trends and Outliers :** Visualizations make it easier to spot trends, correlations, and outliers that might not be apparent in raw data. This ability to identify patterns helps in making data-driven decisions.
- **Enhances Communication :** Visualizations make it easier to communicate data insights to both technical and non-technical audiences. Compelling visuals can tell a story, making it more engaging and memorable.
- **Supports Decision Making :** Data-driven decision-making relies on understanding data, and visualizations provide a clearer understanding of the data’s implications. Whether in business, research, or policymaking, visualizations help in making informed choices.

Overall, data visualization is a crucial tool in the data analysis process. It transforms data into actionable insights, empowers decision-makers, and makes data more accessible and understandable to a broader audience. With the growing availability of powerful data visualization tools and libraries, visualizing data has become an essential skill for anyone working with data in various fields, from business and science to academia and government.

<b>Popular Python Libraries for Data Visualization</b>

- **Matplotlib:** Matplotlib is one of the oldest and most widely used data visualization libraries in Python. It provides a flexible and highly customizable interface for creating a wide range of static, interactive, and animated visualizations.
- **Seaborn:** Seaborn is built on top of Matplotlib and offers a higher-level interface for creating attractive and informative statistical graphics. It simplifies the creation of complex visualizations like heatmaps, violin plots, and pair plots.
- **Pandas:** Pandas, primarily a data manipulation library, also offers built-in visualization capabilities. It allows users to create basic visualizations directly from DataFrame and Series objects.
- **Plotly:** Plotly is a popular library for creating interactive visualizations. It supports a variety of chart types and is well-suited for creating dashboards and web applications.
- **Bokeh:** Bokeh is another excellent library for building interactive visualizations, particularly for web applications. It focuses on providing a concise and powerful API for linking visualizations with Python callbacks.

##### <b>The Data Visualization Process</b>

The data visualization process involves a systematic approach to transform raw data into insightful and visually appealing representations. The process begins with data preparation, where data is cleaned and preprocessed to ensure its suitability for visualization. The next step is to choose the appropriate visualization type, such as line plots, bar charts, scatter plots, or heatmaps, that best represents the data and conveys the intended insights. Adding context through labels, titles, and annotations enhances the visualization’s clarity and understanding.

Selecting appropriate colors and handling scale and axis limits ensures that the visualizations are accurate and avoid misinterpretation. For interactive visualizations, elements like tooltips and zooming can be incorporated to enable users to explore the data further. Regular testing and iteration based on feedback help refine the visualizations and make them more accessible to a broader audience. By following this structured process, data visualization enables users to unlock valuable insights, tell compelling data-driven stories, and make informed decisions.

**Data Visualization Examples**

- Line plots are ideal for visualizing trends and changes over time or continuous data.
- Bar charts are suitable for comparing discrete categories or showing the distribution of a categorical variable.
- Scatter plots are excellent for exploring relationships between two numerical variables and identifying correlations.
- Heatmaps are useful for visualizing the magnitude of a relationship between two categorical variables.
- Pie charts are best for displaying parts of a whole and the proportion of different categories in a dataset.

### **What is Matplotlib and Why you need it?**

Matplotlib library is a data visualization library in Python that allows you to work with data at several levels:

- using the Pyplot module, which treats the plot as a whole;
- through an object-oriented interface, when each shape or part of it is a separate object, this allows you to selectively change their properties and display.

In this article, we will work with the Pyplot module, which is sufficient for plotting.

Matplotlib is used to visualize data of any complexity. The library allows you to build different types of graphs: linear, three-dimensional, scatterplots, and others, as well as combine them.

Additional libraries allow you to expand the possibilities of data analysis. For example, the Cartopy module adds the ability to work with cartographic information. You can learn more about Matplotlib from the [official documentation](https://matplotlib.org/3.6.2/users/index.html).

Matplotlib itself is the basis for other libraries — for example, Seaborn makes it easier to create plots and has more options for visually improving their appearance. But Matplotlib is a basic data visualization library, indispensable in data analysis.

### **How are Matplotlib, Pyplot, and Pylab related?**

When diving into Matplotlib, two modules are mentioned — Pyplot and Pylab. It’s important to understand which one to use in your work and why they appeared. Let’s understand the terminology.

The Matplotlib library is a package for data visualization in Python. Pyplot is a module in the Matplotlib package. You will often see it in code as matplotlib.pyplot. The module helps to automatically create axes, figures and other components without thinking about how this happens. It is Pyplot that is used in most cases.

Pylab is another module that is installed with the Matplotlib package. It simultaneously imports Pyplot and the NumPy library to interactively work with arrays or access plotting functions when working with data.

Now Pylab has only historical significance — it facilitated the transition from MATLAB to Matplotlib, as it allowed you to do without import statements (which is how MATLAB users used to work). You may encounter Pylab in code examples on various sites, but in practice, you won’t need to use the module.

### **Building the first plot in Matplotlib**

First, let’s create two variables — `x` and `y`, which will contain the coordinates of the points along the x and y axes:

```
x = [1, 2, 3, 4, 5]
y = [24, 29, 33, 19, 24]
```

Now let’s build a graph that will connect these points:

```
plt.plot(x, y)
plt.show()
```

![](https://miro.medium.com/v2/resize:fit:788/1*zDxJZ3aWyD3BDqOWeGhUpQ.png)

We got a regular line graph. Let’s break down each command:

- `plt.plot()` is a standard function that draws a plot according to the values that have been passed to it. We passed the coordinates of the points to it;
- `plt.show()` is a function that is responsible for displaying the rendered data on the screen.

Let’s supplement our first chart with a title and labels for the axes:

```
plt.plot(x, y)
plt.xlabel('x-axis') #name of the x-axis
plt.ylabel('y-axis') #name of the y-axis
plt.title('First graph') #name of the graph
plt.show()
```
Let’s take a look at the result:

![](https://miro.medium.com/v2/resize:fit:788/1*rBMifXRwSk4fwWhfQw3kGw.png)

Using Matplotlib, you can customize the display of any plot. For example, we can change the color of the line, as well as highlight the points whose coordinates we set in variables:

```
plt.plot(x, y, color='purple', marker='o', markersize=8)
```

Now the points are clearly visible, and the color of the points and the line has changed to purple:

![](https://miro.medium.com/v2/resize:fit:788/1*HTXlEu5GG52LNRHta3S2Pg.png)

Details about setting the `plt.plot()` function parameters can be found in the [official documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html).

### **Building a Barchart**

This type of visualization allows you to conveniently compare the values of individual variables. In a column chart, the length of the columns is proportional to the measures they represent. As a rule, one of the axes corresponds to one category, and the second corresponds to its discrete value.

For example, a bar chart allows you to visually show the amount of profit by month. Let’s build the following chart:

```
x = ['January', 'February', 'March', 'April', 'May']
y = [2, 4, 3, 1, 7]

plt.bar(x, y, color='purple', label='Profit amount') #label parameter allows you to set the name of the value for the legend
plt.xlabel('Month')
plt.ylabel('$ Profit (in million)')
plt.title('Barchart Example')
plt.legend()
plt.show()
```

The bar chart allows you to see the dynamics of profit changes by months:

### **Combining different chart options**

For some tasks, it’s useful to combine several types of plots, such as a bar chart and a line chart. Let’s refine it by adding dots with profit values to the columns, and connecting them:

```
x = ['January', 'February', 'March', 'April', 'May']
y = [2, 4, 3, 1, 7]

plt.bar(x, y, color='purple', label='Profit amount') #label parameter allows you to set the name of the value for the legend
```

### **Building a Pie Chart**

A pie chart is used to display the composition of groups. For example, we can visually show which brands of cars prevail on the roads of the city. Let’s also play a bit with colors!

```
values = [23, 16, 55, 19, 37]
labels = ["Honda", "Mazda", "BMW", "Mercedes", "Acura"]
colors = ( "#9E70F1", "#F52B94", "#FFE174", "#65E9EB", '#F3907F') #identify colors

plt.pie(values, labels=labels, colors=colors)
plt.title("Distribution of Car Brands on the Road")
plt.show()
```

Let’s take a look at the result:

![](https://miro.medium.com/v2/resize:fit:788/1*kP_xsJts9_JaoVureoIqrQ.png)

It looks good, but it’s unclear what share falls on each car brand. Therefore, it’s always better to supplement pie charts with percentage values. Let’s edit our code by adding the `autopct` parameter to the pie function:

```
values = [23, 16, 55, 19, 37]
labels = ["Honda", "Mazda", "BMW", "Mercedes", "Acura"]
colors = ( "#9E70F1", "#F52B94", "#FFE174", "#65E9EB", '#F3907F') #identify colors

plt.pie(values, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title("Distribution of Car Brands on the Road")
plt.show()
```

In the parameter, we pass the display format of the number. In our case, this will be an integer with one decimal place. Let’s run the code and see the result:

![](https://miro.medium.com/v2/resize:fit:788/1*5kJjRoj5gZMXCpWun_L9uw.png)

Now it’s easier to compare categories as we see numerical values.

## **Seaborn**

I will discuss frequently used plots in exploratory data analysis such as scatterplot, bar plot, box plot and histogram using Seaborn library.

Dataset — [Brain Stroke Dataset from Kaggle](https://www.kaggle.com/datasets/jillanisofttech/brain-stroke-dataset)

Resource — [Seaborn Documentation](https://seaborn.pydata.org/tutorial.html#)

First, we will import the libraries

````python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
````

**Loading the data**

````python
df = pd.read_csv("brain_stroke.csv")
df.head(n=10)
````

### **1. Creating a scatterplot**

- **x** : variable on the x-axis
- **y** : variable on the y-axis
- **data** : dataframe
- **hue** : variable to be used for color
- **style** : variable to be used for marker style
- **legend** : control legend position

**Scatterplot**

````python
sns.scatterplot(x = 'age', y = 'avg_glucose_level', data = df, color = 'magenta')
````

**Scatterplot with hue, style and legend adjustment**

````python
sns.scatterplot(x = 'age', y = 'avg_glucose_level', data = df, alpha = 0.2, hue='gender', style='work_type')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
````

### **2. Creating a barplot**

- **x** : categorical variable
- **y** : numeric value (***mean is calculated by default***)
- **data** : dataframe
- **color** : bar color
- **hue** : variable to be used for color

````python
sns.barplot(x="work_type", y="bmi", data=df, color="skyblue")
````

Horizontal barplot with x and y axis changed and orient = ‘h

````python
sns.barplot(x="bmi", y="work_type", data=df, color="blue", orient="h")
````

**Barplot with variable as color using hue**


````python
sns.barplot(x="work_type", y="bmi", data=df, hue="gender")
````

### **Barplot with label and title**

- plt.xlabel — text for xlable
- plt.ylable — text for y lable
- plt.title — text for title

````python
sns.barplot(x="gender", y="hypertension", data=df, hue="smoking_status")
plt.legend(bbox_to_anchor=(1.4, 1), loc="upper right", title="Smoking Status")
plt.xlabel("Gender")
plt.title("Hypertension % in Gender by Smoking Status")
````

### **Plotting barplot with summary other than mean**

- Summarise the data using groupby
- plot the sumamrised data

````python
df_summ = (
    df.groupby(["work_type", "gender"], as_index=False)["avg_glucose_level"]
    .agg("sum")
    .sort_values(by="avg_glucose_level", ascending=False)
)
sns.barplot(x="work_type", y="avg_glucose_level", data=df_summ, hue="gender")
plt.title("Sum of avg_glucose_level by work_type and gender")
````

### **3. Creating a box plot**

- x : categorical variable
- y : numeric variable
- kind : ‘box’, ‘swarm’ etc.
- data : dataframe

````python
sns.catplot(x="hypertension", y="avg_glucose_level", kind="box", data=df)
plt.title("Distribution of avg_glucose_level by hypertension")
````

### **Box plot with facets**

- **col** : facets in columns
- **row** : facets in rows

````python
sns.catplot(
    x="hypertension",
    y="avg_glucose_level",
    kind="box",
    data=df,
    col="stroke",
    row="gender",
)
````

### **4. Histogram**

- sns.histplot(data = df, x = ‘variable’, bins = n)
- sns.displot(data = df, x = ‘variable’, kind = ‘hist’, bins = n)
- hue : variable for color
- binwidth : width of the bin (either bins or binwidth is to be used)

````python
sns.histplot(data=df, x="avg_glucose_level", bins=20)
````

Histogram using sns.displot

````python
sns.displot(data=df, x="avg_glucose_level", kind="hist", bins=20)
````

Histogram with variable as color using hue

````python
sns.displot(data=df, x="avg_glucose_level", kind="hist", hue="stroke", alpha=0.3)
````

### **5. Density plot**

- sns.kdeplot(data = df, x = ‘variable’)
- sns.displot(data = df, x = ‘variable’, kind = ‘kde’)
- hue : variable for color

````python
sns.displot(data=df, x="avg_glucose_level", kind="kde", hue="gender", alpha=0.5)
````

### **6. Jointplot**

Scatterplot and histogram both in the same plot

````python
sns.jointplot(data=df, x="avg_glucose_level", y="age")
````

### **7. Pairplot**

Scatterplots and individual distribution for all the numerical columns in the same plot

````python
sns.pairplot(data=df)
````

## **Plotly**

**Plotly** is an open-source module of Python which is used for data visualization and supports various graphs like line charts, scatter plots, bar charts, histograms, area plot, etc. In this article, we will see how to plot a basic chart with plotly and also how to make a plot interactive. But before starting you might be wondering why there is a need to learn plotly, so let’s have a look at it.

### Why Plotly

Plotly uses javascript behind the scenes and is used to make interactive plots where we can zoom in on the graph or add additional information like data on hover and many more things. Let’s see few more advantages of plotly –

- Plotly has hover tool capabilities that allow us to detect any outliers or anomalies in a large number of data points.
- It is visually attractive that can be accepted by a wide range of audiences.
- It allows us for the endless customization of our graphs that makes our plot more meaningful and understandable for others.

### Installation

Plotly does not come built-in with Python. To install it type the below command in the terminal.

```python
pip install plotly
```

#### Overview of Plotly Package Structure

in Plotly, there are three main modules –

- **plotly.plotly** acts as the interface between the local machine and Plotly. It contains functions that require a response from Plotly’s server.
- **plotly.graph_objects** module contains the objects (Figure, layout, data, and the definition of the plots like scatter plot, line chart) that are responsible for creating the plots. The Figure can be represented either as dict or instances of plotly.graph_objects.Figure and these are serialized as JSON before it gets passed to plotly.js. Figures are represented as trees where the root node has three top layer attributes – **data, layout,** and **frames** and the named nodes called ‘**attributes**’.

**Note:** plotly.express module can create the entire Figure at once. It uses the graph_objects internally and returns the graph_objects.Figure instance.

**Example:**

```python
import plotly.express as px

# Creating the Figure instance
fig = px.line(x=[1, 2], y=[3, 4])

# printing the figure instance
print(fig)

```

**Output:**

![](https://media.geeksforgeeks.org/wp-content/uploads/20210615214000/plotlyfigureclass.png)

- **plotly.tools** module contains various tools in the forms of the functions that can enhance the Plotly experience.

After going through the basics of plotly let’s see how to create some basic charts using plotly.

### **Line chart**

A **line chart** is one of the simple plots where a line is drawn to shoe relation between the X-axis and Y-axis. It can be created using the **px.line** method with each data position is represented as a vertex  (which location is given by the x and y columns) of a polyline mark in 2D space.

**Syntax:**

```python
Syntax: plotly.express.line(data_frame=None, x=None, y=None, line_group=None, color=None, line_dash=None, hover_name=None, hover_data=None, title=None, template=None, width=None, height=None)

```

**Example:**

```python
import plotly.express as px

# using the iris dataset
df = px.data.iris()

# plotting the line chart
fig = px.line(df, y="sepal_width",)

# showing the plot
fig.show()

```

---

**Output:**

![](https://media.geeksforgeeks.org/wp-content/uploads/20210615215538/linechartplotly.gif)

In the above example, we can see that –

- The labels to the x-axis and y-axis have given automatically by plotly.
- The data of the x-axis and y-axis is shown on hover.
- We can also select a part of the data according to our needs and can also zoom out.
- Plotly also provides a set of tools (seen on the top right corner) to interact with every chart.
- Plotly also allows us to save the graph locally in a static format.

````python
import plotly.express as px

# using the iris dataset
df = px.data.iris()

# plotting the line chart
fig = px.line(
    df,
    y="sepal_width",
)

# showing the plot
fig.show()
````
